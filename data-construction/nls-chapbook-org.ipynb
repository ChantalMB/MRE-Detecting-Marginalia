```python
import os

rootdir = '/Users/cbrou/Desktop/MRE/case-study/nls-data-chapbooks'

book_dirs = []
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        check_sd = len(subdir.split('/'))
        if check_sd == 8:
            book_dirs.append(subdir)

book_dirs = set(book_dirs)
```

```python
for d in book_dirs:
    book_ref = d.split("/")[7]
    img_dir = d + "/image"
    for filename in os.listdir(img_dir):
        fp = os.path.join(img_dir, filename)

        # adding book ref to name of images so they can be linked back to metadata with more ease
        updated_fp = "/Users/cbrou/Desktop/MRE/case-study/nls-chapbook-imgs/" + book_ref + "_" + filename
        os.rename(fp, updated_fp)
```
